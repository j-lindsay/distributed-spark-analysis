```python
import matplotlib.pyplot as plt
import numpy as np

x = []
y = []
colors = []

colorMap = ['r','b','g','c','m','y','k','0.6','0.3','1']
    
cmap = plt.get_cmap('Paired', 10)

with open("/Users/keegan/distributed-spark-analysis/out/nutrients7/part-00000") as f:
    pointsIntermediate = [line.strip("\n") for line in f]
    
    for point in pointsIntermediate:
        name,pointStr,cluster = point.split("$,$,")
        points = pointStr.strip("][").split(",")

        colors.append(int(cluster))
        x.append( float(points[0]) )
        y.append( float(points[1]) )
```

```python
plt.figure(figsize=(20,20))
plt.rcParams['axes.facecolor'] = "white"
plt.scatter(x, y, s=.2, c=colors, cmap=cmap)
```
