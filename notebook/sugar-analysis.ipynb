## Overview


```python
import matplotlib.pyplot as plt

import seaborn as sns
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
data = pd.read_csv('../data/out/user-sugar-intake', sep=',', names=['id', 'sugar']).astype(np.float)
data.head(10)
```

```python
data.describe()
```

```python
product_avg = 13.8453498
user_avg = np.mean(data.sugar)

sns.set(style="whitegrid"); plt.figure(figsize=(16, 8))
plt.title('User / Product Sugar Consumption ')

sns.distplot(data.sugar, bins=300, kde=False).set(xlim=(-1, 50))
plt.axvline(x=user_avg, color='black', linestyle=':',
            label='User Average: {0:.2f}g'.format(user_avg))
plt.axvline(x=product_avg, color='coral', linestyle='--',
            label='Product Average: {0:.2f}g'.format(product_avg))
plt.xlabel('sugar (grams)'); plt.ylabel('users'); plt.legend()
plt.savefig('sugar.png', format='png', dpi=800);
```
