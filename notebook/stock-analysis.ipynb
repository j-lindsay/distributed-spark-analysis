## Overview


```python
import matplotlib.pyplot as plt

import seaborn as sns
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
data = pd.read_csv('../data/out/user-sugar-intake', sep=',', names=['id', 'sugar']).astype(np.float)
data.head(10)
```

```python
data.describe()
```

```python
print('Year Statistics - mean: {0:.2f}, median {1:.2f}'.format(np.mean(data.sugar),
                                                       np.median(data.sugar)))

sns.set(style="whitegrid"); plt.figure(figsize=(18, 8))
plt.title('User Product Sugar Consumption ')

sns.distplot(data.sugar, bins=200, kde=False);
```
